## Lab Parameters

```python
data_path = '/Users/Zhang/notebooks/Amgen DS Challenge/'
image_path = '/Users/Zhang/notebooks/Amgen DS Challenge/figures/'
```

```python
labs_train = pd.read_csv(data_path + 'Train/labs.csv')
```

```python
labs_test = pd.read_csv(data_path + 'Test/labs.csv')
```

```python
labs_train = labs_train.assign(split='1') 
```

```python
labs_test = labs_test.assign(split='0') 
```

```python
labs = labs_train.append(labs_test)
```

```python
len(labs)
```

```python
labs.head()
```

```python
## each patient id has one row
```

```python
dic={'day':"d",'hour':"h"}
```

```python
labs['timestamp'] =labs['Time of Measure (First Day or First Hour of Stay)'].apply(lambda x:dic[x])
```

```python
labs = labs[['patient_id','Lab','Min Value','Max Value','Unit of Measure','timestamp']]
```

```python
labs.columns = ['patient_id','Lab','Min','Max','Unit_of_Measure','timestamp']
```

```python
labs_min = labs.pivot(index='patient_id', columns=['timestamp','Lab'], values='Min')
```

```python
labs_max = labs.pivot(index='patient_id', columns=['timestamp','Lab'], values='Max')
```

```python
labs_min.columns = ["_".join((j,k)) + '_min' for j,k in labs_min.columns]
labs_min.reset_index()
```

```python
labs_max.columns = ["_".join((j,k)) + '_max' for j,k in labs_max.columns]
labs_max.reset_index()
```

```python
labs = labs_min.merge(labs_max, on = 'patient_id', how = 'outer').reset_index()
```

```python
labs.shape
```

```python
labs.columns
```

```python
labs.drop_duplicates(inplace=True)
```

## Merge All Features

```python
features3 = features3.set_index('patient_id')
vitals = vitals.set_index('patient_id')
labs = labs.set_index('patient_id')
```

```python
vitals = vitals.loc[~vitals.index.duplicated(keep='first')]
features3 = features3.loc[~features3.index.duplicated(keep='first')]
labs = labs.loc[~labs.index.duplicated(keep='first')]

df1 = pd.concat([vitals, features3], axis=1)
df_merged = pd.concat([df1, labs], axis=1)
```
